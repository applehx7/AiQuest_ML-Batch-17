```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MaxAbsScaler
```

```python
df = pd.read_csv('supershops - supershops.csv')
```

```python
df.head()
```

```python
df_mksd = df[['Marketing Spend']]
normalization_scaler = MinMaxScaler().fit(df_mksd)
standardization_scaler = StandardScaler().fit(df_mksd)
log_transformer = FunctionTransformer(np.log1p).fit(df_mksd)
rb_scaler = RobustScaler().fit(df_mksd)
ma_scaler = MaxAbsScaler().fit(df_mksd)
```

```python
normalized_value = normalization_scaler.transform(df_mksd)
standardized_value = standardization_scaler.transform(df_mksd)
logt_value = log_transformer.transform(df_mksd)
rbs_value = rb_scaler.transform(df_mksd)
mas_value = ma_scaler.transform(df_mksd)
```

```python
result = pd.concat([df_mksd, pd.DataFrame({'Normalization': normalized_value[:,0], 'Standardization': standardized_value[:,0], 'Log Transformation': logt_value['Marketing Spend'], 'Robust Scaler': rbs_value[:,0], 'Max Asolute Scaler': mas_value[:,0]})], axis=1)
```
